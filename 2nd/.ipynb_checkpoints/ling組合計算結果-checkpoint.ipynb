```python
from elasticsearch import Elasticsearch
import json


def create_index(es):
    body = dict()
    body['settings'] = get_setting()
    body['mappings'] = get_mappings()
    print(json.dumps(body)) #可以用json.dumps輸出來看格式有沒又包錯
    es.indices.create(index='ling組合計算結果', body=body)


def get_setting():
    settings = {
        "index": {
            "number_of_shards": 3,
            "number_of_replicas": 2
        }
    }

    return settings


def get_mappings():
    mappings = {
        "properties": {
            "Number": {
                "type": "integer"
            },
            "A1": {
                "type": "float"
            },
            "A2": {
                "type": "float"
            },
            "A3": {
                "type": "float"
            },
            "A4": {
                "type": "float"
            },
            "A5": {
                "type": "float"
            },
            "A6": {
                "type": "float"
            },
            "A7": {
                "type": "float"
            },
            "A8": {
                "type": "float"
            },
            "A9": {
                "type": "float"
            },
            "Sum": {
                "type": "float"
            },
            "Average": {
                "type": "float"
            }
        }
    }

    return mappings


if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    create_index(es)
```
