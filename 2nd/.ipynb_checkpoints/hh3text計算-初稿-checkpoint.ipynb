```python
from elasticsearch import Elasticsearch
import json

def create_index(es):
    body = dict()
    body['settings'] = get_setting()
    body['mappings'] = get_mappings()
    print(json.dumps(body)) #可以用json.dumps輸出來看格式有沒又包錯
    es.indices.create(index='hh3text', body=body)

def get_setting():
    settings = {
        "index": {
            "number_of_shards": 3,
            "number_of_replicas": 2
        }
    }
    return settings

def get_mappings():
    mappings = {
        "properties": {
            #編號
            "Number": {
                "type": "integer"
            },
            #編號
            "name": {
                "type": "keyword"
            },
            #影片類別
            "subject": {
                "type": "keyword"
            },
            #平均
            "average": {
                "type": "float"
            },
            #一階差
            "first order difference": {
                "type": "float"
            },
            #二階差
            "second order difference": {
                "type": "float"
            },
            #最大心率
            "max heart": {
                "type": "integer"
            },
            #最小心率
            "min heart": {
                "type": "integer"
            },
            #心率變化範圍
            "max to min scope heart": {
                "type": "integer"
            },
            #資訊熵
            "entropy of information": {
                "type": "float"
            },
            #均方根
            "root mean square": {
                "type": "float"
            },
            #標準化均值
            "standardize average": {
                "type": "float"
            }
        }
    }
    return mappings

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    create_index(es)
```

```python
from elasticsearch import Elasticsearch
import json
number=1
name="yu"
naturalsubject="電機及資訊科系測量"
happysubject="電機及資訊科系測量"
sadsubject="設計群測量"
def get_querynatural():
    querynatural = {
        "query": {
            "bool": {
                "must": [{
                    "term": {
                        "name": "yu"
                    }
                },
                    {
                    "term": {
                        "subject.keyword": {"value": naturalsubject}
                    }
                }]
            }
        }
    }
    return querynatural
if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    querynatural = get_querynatural()
    resultnatural = es.search(index='hh3', body=querynatural)
    resultnatural1=json.dumps(resultnatural, ensure_ascii=False)
    resultPulsenatural=json.loads(resultnatural1)
    naturalsize=10
    arrnatural=[0]*naturalsize
    arrnatural1=[0]*naturalsize
    arrnatural2=[0]*naturalsize
    for a in range(0,naturalsize):
        pulse=(resultPulsenatural['hits']['hits'][a]['_source']['pulse'])
        time=(resultPulsenatural['hits']['hits'][a]['_source']['timeForCompare'])        
        print(pulse)
        print(time)
        arrnatural[a]=pulse,time
        print(arrnatural)
    print(sorted(arrnatural,key=lambda s:s[1]))
    arrnatural1=sorted(arrnatural,key=lambda s:s[1])
    print(arrnatural1)
    for b in range(0,10):
        arrnatural2[b]=arrnatural1[b][0]
    print(arrnatural2)

def get_queryhappy():
    queryhappy = {
        "query": {
            "bool": {
                "must": [{
                    "term": {
                        "name": "yu"
                    }
                },
                    {
                    "term": {
                        "subject.keyword": {"value": happysubject}
                    }
                }]
            }
        }
    }
    return queryhappy
if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    queryhappy = get_queryhappy()
    resulthappy = es.search(index='hh3', body=queryhappy)
    resulthappy1=json.dumps(resulthappy, ensure_ascii=False)
    resultPulsehappy=json.loads(resulthappy1)
    happysize=10
    arrhappy=[0]*happysize
    arrhappy1=[0]*happysize
    arrhappy2=[0]*happysize
    for c in range(0,happysize):
        pulse=(resultPulsehappy['hits']['hits'][c]['_source']['pulse'])
        time=(resultPulsehappy['hits']['hits'][c]['_source']['timeForCompare'])        
        print(pulse)
        print(time)
        arrhappy[c]=pulse,time
        print(arrhappy)
    print(sorted(arrhappy,key=lambda s:s[1]))
    arrhappy1=sorted(arrhappy,key=lambda s:s[1])
    print(arrhappy1)
    for d in range(0,10):
        arrhappy2[d]=arrhappy1[d][0]
    print(arrhappy2)

def get_querysad():
    querysad = {
        "query": {
            "bool": {
                "must": [{
                    "term": {
                        "name": "yu"
                    }
                },
                    {
                    "term": {
                        "subject.keyword": {"value": sadsubject}
                    }
                }]
            }
        }
    }
    return querysad
if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    querysad = get_querysad()
    resultsad = es.search(index='hh3', body=querysad)
    resultsad1=json.dumps(resultsad, ensure_ascii=False)
    resultPulsesad=json.loads(resultsad1)
    sadsize=10
    arrsad=[0]*sadsize
    arrsad1=[0]*sadsize
    arrsad2=[0]*sadsize
    for e in range(0,sadsize):
        pulse=(resultPulsesad['hits']['hits'][e]['_source']['pulse'])
        time=(resultPulsesad['hits']['hits'][e]['_source']['timeForCompare'])        
        print(pulse)
        print(time)
        arrsad[e]=pulse,time
        print(arrsad)
    print(sorted(arrsad,key=lambda s:s[1]))
    arrsad1=sorted(arrsad,key=lambda s:s[1])
    print(arrsad1)
    for f in range(0,10):
        arrsad2[f]=arrsad1[f][0]
    print(arrsad2)

    
#正常心率平均
ave = 0.0
for i in range(0,naturalsize):
    ave+=arrnatural2[i]
ave = ave/naturalsize

print("happy心率特徵 :")

#平均
aveB = 0.0
for i in range(0,happysize):
    aveB +=arrhappy2[i]
aveB = aveB/naturalsize
print("平均: ",aveB)

#DiffB1 一階差的平均值
diffB1 = 0.0
meanB1 = 0.0
#X[n-1]-X[n]的總和
for i in range(0,(happysize-1)):
    meanB1+=abs(arrhappy2[i+1]-arrhappy2[i])
diffB1 = meanB1/(naturalsize-1)
print("一階差: ",diffB1)

#DiffB2 二階差的平均值
diffB2 = 0.0
meanB2 = 0.0
#X[n+2]-X[n]的總和
for i in range(0,(happysize-2)):
    meanB2 += abs(arrhappy2[i+2]-arrhappy2[i])
diffB2 = meanB2/(happysize-2)
print("二階差: ",diffB2)

#range 心率變化範圍
scopeB = 0
scopemaxB = 0
scopeminB = 110
for i in range(0,happysize):
    #求最大心率
    if arrhappy2[i]>scopemaxB:
        scopemaxB = arrhappy2[i]
    if arrhappy2[i]<scopeminB:
        scopeminB = arrhappy2[i]
scopeB = scopemaxB-scopeminB
print("最大心率: ",scopemaxB)
print("最小心率: ",scopeminB)
print("心率變化範圍: ",scopeB)

#H(X)資訊熵
x = 0
HB = 0
arrpb = [0.0]*happysize
for i in range(0,happysize):
    x=0 #初值為0
    #心率出現次數
    for j in range(0,happysize):
        if arrhappy2[i]==arrhappy2[j]:
            x+=1
    #p[i]機率
    arrpb[i] = x*1.0/happysize
for k in range(0,happysize):
    HB += -arrpb[k]*math.log(arrpb[k],2)
print("資訊熵: ",HB)

#Radrm 均方根
radrmB = 0.0   
meanrB = 0.0
#相鄰值平方差的總和
for i in range(0,(happysize-1)):
    radrmB += pow(arrhappy2[i+1]-arrhappy2[i],2)
meanrB = radrmB/happysize
print("均方根: ",meanrB)

#標準化均值
arrbb=[0.0]*happysize
aveb = 0.0
for i in range(0,happysize):
    arrbb[i] = arrhappy2[i]-ave
    aveb += arrbb[i]
aveb = aveb/happysize
print("標準化均值: ",aveb)

def load_datas():
    datas = list()
    datas.append(
    {
        #匯入編號
        "Number": number,
        #匯入名字
        "name": name,
        #匯入影片類別
        "subject": happysubject,
        #匯入平均
        "average": float(aveB),
        #匯入一階差
        "first order difference":float(diffB1),
        #匯入二階差
        "second order difference":float(diffB2),
        #匯入最大心率
        "max heart":int(scopemaxB),
        #匯入最小心率
        "min heart":int(scopeminB),
        #匯入心率變化範圍
        "max to min scope heart":int(scopeB),
        #匯入資訊熵
        "entropy of information":float(HB),
        #匯入均方根
        "root mean square":float(meanrB),
        #匯入標準化均值
        "standardize average":float(aveb),
    }
    )
    return datas

def create_data(es, datas):
    for data in datas:
        es.index(index='hh3text', body=data)

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    datas = load_datas()
    create_data(es, datas)

print("subject心率特徵: ")

#平均
aveD = 0.0
for i in range(0,sadsize):
    aveD +=arrsad2[i]
aveD = aveD/sadsize
print("平均: ",aveD)

#DiffD1 一階差的平均值
diffD1 = 0.0
meanD1 = 0.0
#X[n+1]-X[n]的總和
for i in range(0,(sadsize-1)):
    meanD1 += abs(arrsad2[i+1]-arrsad2[i])
diffD1 = meanD1/(sadsize-1)
print("一階差: ",diffD1)

#DiffD2 二階差的平均值
diffD2 = 0.0
meanD2 = 0.0
#X[n+2]-X[n]的總和
for i in range(0,(sadsize-2)):
    meanD2 += abs(arrsad2[i+2]-arrsad2[i])
diffD2 = meanD2/(sadsize-2)
print("二階差: ",diffD2)

#range 心率變化範圍
scopeD = 0
scopemaxD = 0
scopeminD = 110
for i in range(0,sadsize):
    #求最大心率
    if arrsad2[i]>scopemaxD:
        scopemaxD = arrsad2[i]
    #求最小心率
    if arrsad2[i]<scopeminD:
        scopeminD = arrsad2[i]
scopeD = scopemaxD-scopeminD
print("最大心率: ",scopemaxD)
print("最小心率: ",scopeminD)
print("心率變化範圍: ",scopeD)

#H(X)資訊熵
x = 0
HD = 0
arrpd = [0.0]*sadsize
for i in range(0,sadsize):
    x = 0 #初值為0 
    #心率出現次數
    for j in range(0,sadsize):
        if arrsad2[i]==arrsad2[j]:
            x+=1
    arrpd[i] = x*1.0/sadsize
for k in range(0,sadsize):
    HD += -arrpd[k]*math.log(arrpd[k],2)
print("資訊熵: ",HD)

#Radrm 均方根
radrmD = 0.0
meanrD = 0.0
#相鄰值平方差的總和
for i in range(0,(sadsize-1)):
    radrmD += pow(arrsad2[i+1]-arrsad2[i],2)
meanrD = radrmD/sadsize
print("均方根: ",meanrD)

#標準化均值
arrdd = [0.0]*sadsize
aved = 0.0
for i in range(0,sadsize):
    arrdd[i] = arrsad2[i] -ave
    aved +=arrdd[i]
aved = aved/sadsize
print("標準化均值: ",aved)
    
#匯入資料
def load_datas():
    datas = list()
    datas.append(
    {
        #匯入編號
        "Number": number,
        #匯入名字
        "name": name,
        #匯入影片類別
        "subject": sadsubject,
        #匯入平均
        "average": float(aveD),
        #匯入一階差
        "first order difference":float(diffD1),
        #匯入二階差
        "second order difference":float(diffD2),
        #匯入最大心率
        "max heart":int(scopemaxD),
        #匯入最小心率
        "min heart":int(scopeminD),
        #匯入心率變化範圍
        "max to min scope heart":int(scopeD),
        #匯入資訊熵
        "entropy of information":float(HD),
        #匯入均方根
        "root mean square":float(meanrD),
        #匯入標準化均值
        "standardize average":float(aved),
    }
    )
    return datas

def create_data(es, datas):
    for data in datas:
        es.index(index='hh3text', body=data)

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    datas = load_datas()
    create_data(es, datas)
```
