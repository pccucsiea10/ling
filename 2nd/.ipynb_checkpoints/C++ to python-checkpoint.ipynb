```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
import operator
import math
```

```python
Max=100 #正常的最大心率
Min=50  #正常的最小心率

A = 5
B = 11
D = 9
arra = [94, 90, 91, 98, 94]
arrb = [85, 87, 92, 100, 93, 93, 100, 94 ,90, 90 ,86]
arrd = [84, 76, 86, 76, 81, 84, 81, 79, 86]

#正常心率平均
ave = 0.0
for i in range(0,A):
    ave+=arra[i]
ave = ave/A

print("A視頻心率特徵 :")

#平均
aveB = 0.0
for i in range(0,B):
    aveB +=arrb[i]
aveB = aveB/B
print("平均: ",aveB)

#Diff1 一階差的平均值
diff1 = 0.0
mean1 = 0.0
#X[n-1]-X[n]的總和
for i in range(0,(B-1)):
    mean1+=abs(arrb[i+1]-arrb[i])
diff1 = mean1/(B-1)
print("一階差: ",diff1)

#Diff2 二階差的平均值
diff2 = 0.0
mean2 = 0.0
#X[n+2]-X[n]的總和
for i in range(0,(B-2)):
    mean2 += abs(arrb[i+2]-arrb[i])
diff2 = mean2/(B-2)
print("二階差: ",diff2)

#range 心率變化範圍
scope = 0
scopemax = 0
scopemin = 110
for i in range(0,B):
    #求最大心率
    if arrb[i]>scopemax:
        scopemax = arrb[i]
    if arrb[i]<scopemin:
        scopemin = arrb[i]
print("最大心率: ",scopemax)
print("最小心率: ",scopemin)
print("心率變化範圍: ",scopemax-scopemin)

#H(X)資訊熵
x = 0
H = 0
arrpb = [0.0]*B
for i in range(0,B):
    x=0 #初值為0
    #心率出現次數
    for j in range(0,B):
        if arrb[i]==arrb[j]:
            x+=1
    #p[i]機率
    arrpb[i] = x*1.0/B
for k in range(0,B):
    H += -arrpb[k]*math.log(arrpb[k],2)
print("資訊熵: ",H)

#Radrm 均方根
radrm = 0.0
meanr = 0.0
#相鄰值平方差的總和
for i in range(0,(B-1)):
    radrm += pow(arrb[i+1]-arrb[i],2)
meanr = radrm/B
print("均方根: ",meanr)

#標準化均值
arrbb=[0.0]*B
aveb = 0.0
for i in range(0,B):
    arrbb[i] = arrb[i]-ave
    aveb += arrbb[i]
aveb = aveb/B
print("標準化均值: ",aveb)

print("C視頻心率特徵: ")

#平均
aveD = 0.0
for i in range(0,D):
    aveD +=arrd[i]
aveD = aveD/D
print("平均: ",aveD)

#Diff1 一階差的平均值
diff1 = 0.0
mean1 = 0.0
#X[n+1]-X[n]的總和
for i in range(0,(D-1)):
    mean1 += abs(arrd[i+1]-arrd[i])
diff1 = mean1/(D-1)
print("一階差: ",diff1)

#Diff2 二階差的平均值
diff2 = 0.0
mean2 = 0.0
#X[n+2]-X[n]的總和
for i in range(0,(D-2)):
    mean2 += abs(arrd[i+2]-arrd[i])
diff2 = mean2/(D-2)
print("二階差: ",diff2)

#range 心率變化範圍
scope = 0
scopemax = 0
scopemin = 110
for i in range(0,D):
    #求最大心率
    if arrd[i]>scopemax:
        scopemax = arrd[i]
    #求最小心率
    if arrd[i]<scopemin:
        scopemin = arrd[i]
print("最大心率: ",scopemax)
print("最小心率: ",scopemin)
print("心率變化範圍: ",scopemax-scopemin)

#H(X)資訊熵
x = 0
H = 0
arrpd = [0.0]*D
for i in range(0,D):
    x = 0 #初值為0 
    #心率出現次數
    for j in range(0,D):
        if arrd[i]==arrd[j]:
            x+=1
    arrpd[i] = x*1.0/D
for k in range(0,D):
    H += -arrpd[k]*math.log(arrpd[k],2)
print("資訊熵: ",H)

#Radrm 均方根
radrm = 0.0
meanr = 0.0
#相鄰值平方差的總和
for i in range(0,(D-1)):
    radrm += pow(arrd[i+1]-arrd[i],2)
meanr = radrm/D
print("均方根: ",meanr)

#標準化均值
arrdd = [0.0]*D
aved = 0.0
for i in range(0,D):
    arrdd[i] = arrd[i] -ave
    aved +=arrdd[i]
aved = aved/D
print("標準化均值: ",aved)
```
