```python
from elasticsearch import Elasticsearch
import json


def create_index(es):
    body = dict()
    body['settings'] = get_setting()
    body['mappings'] = get_mappings()
    print(json.dumps(body)) #可以用json.dumps輸出來看格式有沒又包錯
    es.indices.create(index='1-01st', body=body)


def get_setting():
    settings = {
        "index": {
            "number_of_shards": 3,
            "number_of_replicas": 2
        }
    }

    return settings


def get_mappings():
    mappings = {
        "properties": {
            "name": {
                "type": "integer"
            },
            "feel": {
                "type": "integer"
            },
            "pulse": {
                "type": "integer"
            },
            "time": {
                "type": "integer"
            }
        }
    }

    return mappings


if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    create_index(es)
```

```python
from elasticsearch import Elasticsearch

def load_datas():
    datas = list()
    with open('1-0-1st.csv', 'r') as f:
        for data in f.readlines():
            name, feel, pulse, time = data.replace('\n', '').split(',')
            datas.append(
                {
                    "name": int(name),
                    "feel": int(feel),
                    "pulse": int(pulse),
                    "time": int(time)
                }
            )
    return datas

def create_data(es, datas):
    for data in datas:
        es.index(index='1-01st', body=data)  

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    datas = load_datas()
    create_data(es, datas)
```
