```python
#建立ling-林郁芮數據01資料庫

from elasticsearch import Elasticsearch
import json

def create_index(es):
    body = dict()
    body['settings'] = get_setting()
    body['mappings'] = get_mappings()
    print(json.dumps(body)) #可以用json.dumps輸出來看格式有沒又包錯
    es.indices.create(index='ling-林郁芮數據01', body=body)

def get_setting():
    settings = {
        "index": {
            "number_of_shards": 3,
            "number_of_replicas": 2
        }
    }
    return settings

def get_mappings():
    mappings = {
        "properties": {
            "name": {
                "type": "keyword"
            },
            "feel": {
                "type": "keyword"
            },
            "pulse": {
                "type": "integer"
            },
            "time": {
                "type": "integer"
            },
            "trial":{
                "type":"keyword"
            }
        }
    }
    return mappings

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    create_index(es)
```

```python
#傳送ling-林郁芮natural數據-1st.csv到ling-林郁芮數據01資料庫

from elasticsearch import Elasticsearch

def load_datas():
    datas = list()
    with open('ling-林郁芮natural數據-1st.csv', 'r') as f:
        for data in f.readlines():
            name, feel, pulse, time, trial = data.replace('\n', '').split(',')
            datas.append(
                {
                    "name": name,
                    "feel": feel,
                    "pulse": int(pulse),
                    "time": int(time),
                    "trial": trial
                }
            )
    return datas

def create_data(es, datas):
    for data in datas:
        es.index(index='ling-林郁芮數據01', body=data)  

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    datas = load_datas()
    create_data(es, datas)
```
