```python
#傳送ling-張哲唯natural數據-2nd.csv到ling-張哲唯數據01資料庫

from elasticsearch import Elasticsearch

def load_datas():
    datas = list()
    with open('ling-張哲唯natural數據-2nd.csv', 'r') as f:
        for data in f.readlines():
            name, feel, pulse, time, trial = data.replace('\n', '').split(',')
            datas.append(
                {
                    "name": name,
                    "feel": feel,
                    "pulse": int(pulse),
                    "time": int(time),
                    "trial": trial
                }
            )
    return datas

def create_data(es, datas):
    for data in datas:
        es.index(index='ling-張哲唯數據01', body=data)  

if __name__ == "__main__":
    es = Elasticsearch(hosts='140.137.41.81', port=9200)
    datas = load_datas()
    create_data(es, datas)
```
