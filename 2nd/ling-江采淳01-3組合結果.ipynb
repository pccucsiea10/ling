```python
arra = [0, 0, 0]
statics = [0, 0, 0]
staticsA = [0, 0, 0]
for k in range(0, 8):
    for m in range(0, 3):
        for j in range(0, 3):
            reverse = 2 - j
            if arra[reverse] == 2:
                arra[reverse] = 0
                arra[reverse - 1] += 1
        print(arra[m], end = '')
    print()
    test = []
    if arra[0] == 1:
        if len(test):
            test = test
        else:
            test = np.c_[arrX0]
    if arra[1] == 1:
        if len(test):
            test = np.c_[test, arrX1]
        else:
            test = np.c_[arrX1]
    if arra[2] == 1:
        if len(test):
            test = np.c_[test, arrX2]
        else:
            test = np.c_[arrX2]
    arra[2] += 1
    if k > 0:
        X_train, X_test, y_train, y_test = train_test_split(test, y,test_size=0.3,random_state=1)
        maxScore = 0
        kn = 0
        for o in range(5, 20):
            clf = KNeighborsClassifier(n_neighbors = o,p = 2,weights = 'distance',algorithm = 'auto')
            clf.fit(X_train,y_train)
            score = clf.score(X_test,y_test)
            if score > maxScore:
                maxScore = score
                kn = o
        print(maxScore,kn)
        if maxScore == 1.0 :
            for i in range(0,3):
                if i<2:
                    statics[i] += arra[i]
                else:
                    b=arra[i]-1
                    statics[i] += b
        elif  maxScore >= 0.9:
            for i in range(0,3):
                if i<2:
                    staticsA[i] += arra[i]
                else:
                    b=arra[i]-1
                    staticsA[i] += b
print("總和1.0的=",statics)
print("總和0.9的=",staticsA)
```
